```python
#Import required libraries
import pandas as pd
from sqlalchemy import create_engine
import pymysql
```

```python
#Create engine
engine = create_engine('mysql+pymysql://root:@ip:port/folder')
```

```python
#Connecting engine
conn = engine.connect()
```

```python
#Read a query into a DataFrame
df = pd.read_sql('''SELECT * FROM data1202.vgsales_2016''', conn)
df.head()
```

```python
#Was the average of global sales higher before or after 2005?
before = df[df['Year_of_Release'] < 2005].Global_Sales.mean()
after = df[df['Year_of_Release'] >= 2005].Global_Sales.mean()
print('The average of global sales before 2005 was', round(before, 2), 'while the average after 2005 was', round(after, 2))
```

```python
#Create a new column that labels records before 2005 as 'pre-2005' and after 2005 as 'post-2005'
df['pre_or_post_2005'] = df['Year_of_Release'] < 2005
df['pre_or_post_2005'] = df.pre_or_post_2005.map({True:'pre-2005', False:'post-2005'})

df.head()
```
